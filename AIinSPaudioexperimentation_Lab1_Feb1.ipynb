```python
import librosa
import matplotlib.pyplot as plt
import IPython.display as ipd
from pydub import AudioSegment
from pydub.playback import play
import numpy as np
```

```python
#loading the audio file using librosa
y, sr = librosa.load('/Users/localadmin/Desktop/AIinSPaudiofeb1.wav')
```

```python
#Waveform graph for whole audio file
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y)
plt.title('Waveform - Audio File AI in Speech Processing')
```

```python
#playing the audio 
ipd.Audio(y, rate=sr)
```

```python
#creating a segment from the file
start_time = 500
end_time = 1000

start_frame = int(librosa.time_to_frames(start_time, sr=sr))
end_frame = int(librosa.time_to_frames(end_time, sr=sr))
segment = y[start_frame:end_frame]
```

```python
ipd.Audio(segment, rate=sr)
```

```python
#Waveform graph for small segment from audio file
plt.figure(figsize=(10, 4))
librosa.display.waveshow(segment, sr=sr)
plt.title('Waveform - Small Segment')
```

```python
#reversing the audio
reversed = np.flipud(y)
```

```python
ipd.Audio(reversed, rate=sr)
```

```python
#Waveform graph for reversed audio from audio file
plt.figure(figsize=(10, 4))
librosa.display.waveshow(reversed, sr=sr)
plt.title('Waveform - reversed audio')
```

```python
fastforward = librosa.effects.time_stretch(y, rate=2.0)
```

```python
ipd.Audio(fastforward, rate=sr)
```

```python
slowedversion = librosa.effects.time_stretch(y, rate=0.5)
```

```python
ipd.Audio(slowedversion, rate=sr)
```
