A1.Find the first derivative of your speech signal with finite difference method. Listen to the first derivative signal and the original speech signal.

```python
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np

filename = '/Users/localadmin/Desktop/AISP/AIinSPaudiofeb1.wav'

import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt

speech_signal, sr = sf.read(filename)
time = np.arange(len(speech_signal)) / sr

firstDerivative = np.diff(speech_signal) / np.diff(time)

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time, speech_signal, color='y')
plt.title('Original Speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(time[:-1], firstDerivative, color='pink')
plt.title('First Derivative of Speech Signal')
plt.xlabel('Time (s)')
plt.ylabel('First Derivative')

plt.tight_layout()
plt.show()

sf.write('first_derivative_signal.wav', firstDerivative, sr)
```

```python
ipd.Audio(firstDerivative, rate=sr)
```

A2. Detect the points of zero crossing in the first derivative signal. Compare the average length between two consecutive zero crossings for speech and silence regions. Observe the pattern.

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
import soundfile as sf
speech_derivative, sample_rate = sf.read('first_derivative_signal.wav')

zero_crossings = np.where(np.diff(np.sign(speech_derivative)))[0]

zero_crossing_lengths = np.diff(zero_crossings)
speech_threshold = 0.01  
silence_threshold = 0.001  

speech_regions = zero_crossing_lengths[zero_crossing_lengths > speech_threshold * sample_rate]
silence_regions = zero_crossing_lengths[zero_crossing_lengths <= silence_threshold * sample_rate]

avg_length_speech = np.mean(speech_regions)
avg_length_silence = np.mean(silence_regions)

print("Average length of zero crossings in speech:", avg_length_speech / sample_rate, "seconds")
print("Average length of zero crossings in silence regions:", avg_length_silence / sample_rate, "seconds")
```
